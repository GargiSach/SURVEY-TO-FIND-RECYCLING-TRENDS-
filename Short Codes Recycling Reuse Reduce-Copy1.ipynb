### Approach adopted for building combined subplot

1. Assigned name (combinebarplot) to the function 
2. Inside the function included following 5 attributes :
2a. List 1 - for including names of the dataframes for which subplots was required as output
2b. x
2c. y
2d. List 2 - List of Titles that fuction will assign to each of the subplot 
2e. List 3 - Dictionary containing color code for each class under variable x 
3. Used for loop to get index positon at each iteration
4. The same used for changing subplot number, title for each subplot 
5. Under the fuction gave standard fontsizes for axis, legends and labels

```python
#Function for extracting pie plots for each item's dataframe
def combinepiechart(df,x,y,colors):
    plt.figure(figsize=(9,7))
    values = list(df[y])
    labels = list(colors.keys())
    c = df[x].apply(lambda x:colors[x])
    _,_, autotexts=plt.pie(df[y],colors=c,autopct='%1.1f%%',counterclock=True, shadow=False,labels=values,startangle=90,wedgeprops={"edgecolor":'white','linewidth':2,'linestyle':'dashed','antialiased':True},textprops={'fontsize':15})
    plt.title(f'{x}', fontdict={'fontsize':15,'color':'blue'}, y=1.01)
    for autotext in autotexts:
        autotext.set_color('white')
    plt.legend(labels,loc=4)
    plt.show()
```

```python
#Function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i],fontdict={'fontsize':20,'color':'Black'})
```

```python
#Function for extracting bar plots for each item's dataframe (both at locality and as well as at society's level)
def combinebarplot(lis,x,y,colors,Title):
    plt.figure(figsize=(30,25))

    for i in range (0,len(lis)):
        
        plt.subplot(3,3,i+1)
        labels = list(colors.keys())
        #abels2=['\n'.join(wrap(l,10))for l in labels]
        handles = [plt.Rectangle((0,0),1,1,color=colors[label]) for label in labels]
        c = lis[i][x].apply(lambda x:colors[x])
        bar = plt.bar(lis[i][x],lis[i][y],color=c,edgecolor='black',linewidth=2,label=labels)
        addlabels(lis[i][x],lis[i][y])
        plt.title(f'{Title[i]}', fontdict={'fontsize':15,'color':'blue'}, y=1.01)
        plt.xlabel(f'{x}',fontdict={'fontsize' : 12,'color':'black'})
        plt.ylabel(f'{y}',fontdict={'fontsize' : 14,'color':'black'})
        plt.xticks(fontsize=15)
        plt.yticks(fontsize=15,rotation = 0)
        plt.legend(handles,labels,fontsize=15)
        
    plt.show()
```
